```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.datasets import load_iris
```

```python
#Load the data
data = load_iris().data
```

```python
# Determination of the number of clusters using the Elbow method for unnormalized data
k_range = range(2, 13)
sse = []
 
for k in k_range:
    k_means = KMeans(n_clusters=k, n_init=10)
    k_means.fit_predict(data)
    sse.append(k_means.inertia_)
```

```python
# Visualize the inertia versus the number of clusters for unnormalized data
plt.figure('Elbow Method')
plt.plot(k_range, sse, c='r', marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('SSE')
plt.title('The Elbow Method using Inertia')
plt.grid()
plt.show()
```

```python
#NORMALIZATION with standardScaler:
norm1 = StandardScaler()
data_norm1 = data.copy()
data_norm1 = norm1.fit_transform(data_norm1)
#NORMALIZATION with MinMax:
norm2 = MinMaxScaler()
data_norm2 = data.copy()
data_norm2 = norm2.fit_transform(data_norm2)
```

```python
# Determination of the number of clusters using the Elbow method for normalized data with StandardScaler
k_range = range(2, 13)
sse_norm = []
for k in k_range:
    k_means = KMeans(n_clusters=k, n_init=10)
    k_means.fit_predict(data_norm1)
    sse_norm.append(k_means.inertia_)
```

```python
# Visualize the inertia versus the number of clusters for normalized data with StandardScaler
plt.figure(num='Elbow Method for normalized data with StandardScaler')
plt.plot(k_range, sse_norm, c='g', marker='+')
plt.xlabel('Number of clusters')
plt.ylabel('SSE')
plt.title('The Elbow Method for normalized data with StandardScaler')
plt.grid()
plt.show()
```

```python
# Determination of the number of clusters using the Elbow method for normalized real estate data with MinMaxScaler
k_range = range(2, 13)
sse_norm = []
for k in k_range:
    k_means = KMeans(n_clusters=k, n_init=10)
    k_means.fit_predict(data_norm2)
    sse_norm.append(k_means.inertia_)
```

```python
# Visualize the inertia versus the number of clusters for normalized data with MinMaxScaler
plt.figure(num='Elbow Method for normalized data with MinMax')
plt.plot(k_range, sse_norm, c='m', marker='^')
plt.xlabel('Number of clusters')
plt.ylabel('SSE')
plt.title('The Elbow Method for normalized data with Max')
plt.grid()
plt.show()
```

```python
# From the Elbow Method plot, we choose the optimal number of clusters
optimal_k = 3  
kmeans = KMeans(n_clusters=optimal_k , n_init=10)
```

```python
# Performing clustering with the optimal number of clusters for unnormalized data
cluster_labels = kmeans.fit_predict(data)
print(f"cluser labels for unnormalized data: {cluster_labels}")
```

```python
# Visualizing the clustering results
plt.figure()
plt.scatter(data[:, 0], data[:, 1], c=cluster_labels, cmap='viridis')
plt.title('Clustering Results')
plt.show()
```
